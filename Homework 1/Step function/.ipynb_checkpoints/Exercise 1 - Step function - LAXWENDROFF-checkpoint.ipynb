# Step function - Lax-Wendroff
I solve the the advection equation using as an initial condition a step function $u(x, t = 0) = 1$ for $x \in [4,6]$ and $u(x, t = 0) = 0$ in the rest of the domain. I use a domain $x \in [0,10]$ with periodic boundary conditions.

I use the Lax-Wendroff scheme.

```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
L = 10.0     
nx = 101    
dx = L/(nx-1)   
x = np.linspace(0, L, nx)

cf = 0.5       
dt = cf*dx
t_final = 20.000

u_initial = np.empty(len(x))

for i in range (len(x)):
    if x[i]>=4 and x[i]<=6:
        u_initial[i] = 1
    else:
        u_initial[i] = 0
        
u_current = u_initial.copy()   

l2norm=[]
l2norm.append(np.sqrt(np.sum(u_current**2)/len(u_current)))

time=[]
time.append(0.0)
        
print('nx =',nx)
print('dx =',dx)
print('dt =',dt)
print('Number of iterations =',t_final/dt)
```

```python
os.makedirs('./images_LAXWENDROFF_101points')
os.listdir('./')
```

```python
t = 0.0
i = 0

plt.plot(x, u_current)
plt.title('Time='+str(round(t,2)))
plt.ylim(-0.3,1.3)
plt.savefig('./images_LAXWENDROFF_101points/fig_'+str(i).zfill(5)+'.png', dpi=200)
plt.close()

while t < t_final:
    
    u_next = u_current - dt/(2*dx)*(np.roll(u_current, -1) - np.roll(u_current, 1)) + dt**2/(2*dx**2)*(np.roll(u_current, -1)-2*u_current+np.roll(u_current, 1))
    u_current = u_next.copy()

    t += dt
    i += 1
    
    l2norm.append(np.sqrt(np.sum(u_current**2)/len(u_current)))
    time.append(t)
    
    if (i%10==0):
        plt.plot(x, u_current)
        plt.title('Time = '+str(round(t,2)))
        plt.ylim(-0.3,1.3)
        plt.savefig('./images_LAXWENDROFF_101points/fig_'+str(i).zfill(5)+'.png', dpi=200)
        plt.close()
```

```python
plt.plot(x, u_initial, label='Initial')
plt.plot(x, u_current, label='Final')
plt.ylim(-0.3,1.3)
plt.legend()
plt.savefig('init_final_step_LAXWENDROFF_101points.png', dpi=200)
```

```python
np.savetxt('l2norm_LAXWENDROFF_101points.txt', l2norm)
np.savetxt('time_LAXWENDROFF_101points.txt', time)
```
