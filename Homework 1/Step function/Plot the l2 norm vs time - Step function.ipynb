```python
import matplotlib.pyplot as plt
import numpy as np
```

### Lax-Friedrichs

```python
l2_101 = np.genfromtxt('./l2norm_LAX_101points.txt', comments='#')
l2_501 = np.genfromtxt('./l2norm_LAX_501points.txt', comments='#')
l2_1001 = np.genfromtxt('./l2norm_LAX_1001points.txt', comments='#')
l2_02 = np.genfromtxt('./l2norm_LAX_0.2cf.txt', comments='#')
l2_1 = np.genfromtxt('./l2norm_LAX_1cf.txt', comments='#')
l2_12 = np.genfromtxt('./l2norm_LAX_1.2cf.txt', comments='#')
time_101 = np.genfromtxt('./time_LAX_101points.txt', comments='#')
time_501 = np.genfromtxt('./time_LAX_501points.txt', comments='#')
time_1001 = np.genfromtxt('./time_LAX_1001points.txt', comments='#')
time_02 = np.genfromtxt('./time_LAX_0.2cf.txt', comments='#')
time_1 = np.genfromtxt('./time_LAX_1cf.txt', comments='#')
time_12 = np.genfromtxt('./time_LAX_1.2cf.txt', comments='#')

fig, ax = plt.subplots(1,1, figsize=(8,5))
ax.plot(time_101, l2_101/l2_101[0], alpha = 0.5, label = 'J=101', c='b')
ax.plot(time_501, l2_501/l2_501[0], alpha = 0.5, label = 'J=501', c='orange')
ax.plot(time_1001, l2_1001/l2_1001[0], alpha = 0.5, label = 'J=1001', c='r')
ax.plot(time_02, l2_02/l2_02[0], alpha = 0.5, label = r'$c_f$=0.2', c='g')
ax.plot(time_1, l2_1/l2_1[0], alpha = 0.5, label = r'$c_f$=1')
ax.plot(time_12, l2_12/l2_12[0], alpha = 0.5, label = r'$c_f$=1.2', c='darkmagenta')

ax.legend(loc= 'upper right')
ax.set_ylim (0.4, 1.9)
ax.set_xlabel('Time')
ax.set_ylabel(r'$l_2$ norm')
ax.grid (c='gray', alpha = 0.3)

plt.xticks(visible=True)
plt.draw()

plt.savefig('comparison_l2norm_LAXFRIEDRICHS.png', dpi=200)
```
