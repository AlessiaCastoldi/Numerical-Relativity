```python
import matplotlib.pyplot as plt
import numpy as np
import kuibit.cactus_scalars as cs
import kuibit.simdir as sd
from kuibit import grid_data as gd
```

```python
sim1 = sd.SimDir("./TOV_lev2_r12/")
sim2 = sd.SimDir("./TOV_lev3_r12_r10/")
sim3 = sd.SimDir("./TOV_lev3_r12_r10_res/")
```

```python
timeseries1 = sim1.timeseries
timeseries2 = sim2.timeseries
timeseries3 = sim3.timeseries
```

```python
rho_max1=timeseries1.maximum.fields.rho
rho_max2=timeseries2.maximum.fields.rho
rho_max3=timeseries3.maximum.fields.rho
```

```python
plt.plot(rho_max1, label="Two grids with dx=2.0")
plt.plot(rho_max2, label="Three grids with dx=2.0")
plt.plot(rho_max3, label="Three grids  with dx=1.5")
plt.xlabel(r'$t$ [$M_{\odot}$]')
plt.ylabel(r'$\rho_c$')
plt.legend()
plt.savefig('TOV_rhomax.png', dpi=300)
```

```python
gf = sim3.gf
```

```python
vars2D = gf.xy
```

```python
rho = vars2D.fields.rho
```

```python
small_grid = gd.UniformGrid([100, 100], x0=[0, 0], x1=[18,18])
rho_small = rho.read_on_grid(544768, small_grid)

cf = plt.pcolormesh(*rho_small.coordinates_meshgrid(), np.log10(rho_small.data_xyz), vmin=-10, vmax=-3)
plt.colorbar(cf)

plt.savefig ('TOV_lev3_r12_r10_res_last_iteration.png', dpi=300)
```

```python
small_grid = gd.UniformGrid([100, 100], x0=[0, 0], x1=[18,18])
rho_small = rho.read_on_grid(0, small_grid)

cf = plt.pcolormesh(*rho_small.coordinates_meshgrid(), np.log10(rho_small.data_xyz), vmin=-10, vmax=-3)
plt.colorbar(cf)

plt.savefig ('TOV_lev3_r12_r10_res_first_iteration.png', dpi=300)
```
